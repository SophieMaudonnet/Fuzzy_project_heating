# Test for getting values from sensor

In "fuzzy" conda environment, import all libraries for fuzzy course + pyserial <br>
Find port with "ls /dev/tty.*" <br>
See https://randomnerdtutorials.com/bme280-sensor-arduino-pressure-temperature-humidity/

```python
import serial
import re
```

```python
ser = serial.Serial('COM3')

line = ser.readline().decode()

while 'Temperature' not in line:
    line = ser.readline().decode()
temp = line

while 'Pressure' not in line:
    line = ser.readline().decode()
pres = line

while 'Altitude' not in line:
    line = ser.readline().decode()
alt = line

while 'Humidity' not in line:
    line = ser.readline().decode()
humid = line

ser.close()
```

```python
humid
```

```python
def extract_data(output):
    data = re.findall(r'\d+\.\d+', output)[0]
    return float(data)
```

```python
t = extract_data(temp)
p = extract_data(pres)
a = extract_data(alt)
h = extract_data(humid)

print(f'Temperature: {t}°C, Pressure: {p}hPa, Altitude: {a}m, Humidity: {h}%')
```
